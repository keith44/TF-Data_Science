DRILL - Descriptive Statistics and Normality

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beta = np.random.beta(3.5, 1.6, 100)
plt.hist(beta, bins=20)
plt.show()

#Plot a histogram for beta.
plt.hist(beta, bins=20, color='blue')

# Add a vertical line at the mean.
plt.axvline(beta.mean(), color='r', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(beta.mean() + beta.std(), color='r', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(beta.mean()-beta.std(), color='r', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

The data here is skewed left a little meaning the majority of the data is on the right.

```python
logistic  = np.random.logistic(0, 1, 100)
plt.hist(logistic, bins=20)
plt.show()

 
plt.hist(logistic, bins=20, color='blue')
plt.axvline(logistic.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(logistic.mean() + beta.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(logistic.mean()-beta.std(), color='r', linestyle='dashed', linewidth=2) 
plt.show()
```

It seems the data here is close to normal and the mean and standard deviation are good measures of central tendancy and variance.

DRILL - Exploring the Central Limit Theorem

```python
# I would expect as the sample size gets larger the mean and std dev will get closer to the values of the population.


pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000)

print("mean1: " + str (pop1.mean()))
print("std dev1: " + str (pop1.std()) + "\n")

print("mean2: " + str (pop2.mean()))
print("std dev2: " + str (pop2.std()))
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

print("mean1: " + str (sample1.mean()))
print("std dev1: " + str (sample1.std()) + "\n")

print("mean2: " + str (sample2.mean()))
print("std dev2: " + str (sample2.std()))
```

```python
# #1. Increase sample size from 100 to 1000

sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print("mean1: " + str (sample1.mean()))
print("std dev1: " + str (sample1.std()) + "\n")

print("mean2: " + str (sample2.mean()))
print("std dev2: " + str (sample2.std()))
```

```python
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print("mean1: " + str (sample1.mean()))
print("std dev1: " + str (sample1.std()) + "\n")

print("mean2: " + str (sample2.mean()))
print("std dev2: " + str (sample2.std()))
```

There is little change between the standard deviations and the means of all three samples. There is a rather big difference between the means of sample 1 and sample 2.

```python
# #2. increase pop1 p-value to .3

pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
# increase pop1 p-value to .4

pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

The Ttest gets smaller and the p-value gets larger as the sample p-values get closer to the same value. The larger the p-value the more likely there is no difference between the means of the two groups. The t-value gets smaller because the smaller the difference of the means is relative to the variance, the more likely the difference is not due to random chance, and the more likely it is to reflect similarities between the two populations.

```python
# #3. 

geometric1 = np.random.geometric(.1,10000)
geometric2 = np.random.geometric(.5,10000)
sample1 = np.random.choice(geometric1, 1000, replace=True)
sample2 = np.random.choice(geometric2, 1000, replace=True)

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

Yes. It appears that the sample means represent the population.
