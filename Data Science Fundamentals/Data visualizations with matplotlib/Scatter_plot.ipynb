```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/keith/Desktop/titanic.csv')
```

```python
a = df[['Fare']]
b = df[["Age"]].sort_values('Age')
```

```python
plt.scatter(
    x=b,
    y=a,
    color='green',
    marker='x', s=10
)
plt.ylabel('Fare')
plt.xlabel('Age')
plt.title('Fare vs. Age')
plt.show()
```
